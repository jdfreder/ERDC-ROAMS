```python


grid = GridWidget(2,1)
grid.height=870
display(grid)

time_slider = FloatSliderWidget(description="Time (day):", step=0.01)
#time_slider.set_css('width', '80%')
time_slider_container=  ContainerWidget()
tides = TideChart(time_slider_container)
time_slider_container.children=time_slider_container.children + tuple([time_slider])
time_slider_container.set_css({
    'margin-top': '40px',
    'width': '100%',
    'text-align': 'center',
})

grid[1,0].children = []
bath = BathymetryChart(grid[1,0])
currents = CurrentChart(grid[1,0])
waves = WavesChart(grid[1,0])
#chart3 = labeled_chart(grid[1,0], 'Wave Height', 'Height (m)', 'Distance (m)')
grid[1,0].children = grid[1,0].children
grid[1,0].set_css('text-align', 'center')
display(time_slider_container
        )
time_slider_container.add_class('center align-center')

    
def render_frame(time=0., update=False):
    ####### CHART 1 DATA ################################################
    ballast_depth = -6
    data = np.array([(i, -5.*sin(float(time+i)/8.)*sin(float(time+i)/10.)*sin(float(time+i)/13.)-15.) for i in range(200)])
    uncert = np.array([(i, 2.) for i in range(200)])
    #####################################################################
    bath.fill(ballast_depth, data, uncert, update=update)

    ####### CHART 2 DATA ################################################
    data = np.array([(i, 5.*sin(float(time+i)/10.)*sin(float(time+i)/11.)*sin(float(time+i)/14.)) for i in range(200)])
    #####################################################################
    currents.fill(data, update=update)

    ####### CHART 3 DATA ################################################
    data = np.array([(i, .2*sin(float(time+i)/20.)*sin(float(time+i)/31.)*sin(float(time+i)/13.)+.2) for i in range(200)])
    #####################################################################
    waves.fill(data, update=update)
    
    tides.fill(time, update=update)

time_slider.min = 0.
time_slider.max = 12. * 12. / 24.
render_frame(0.)
def time_slide(name, new):
    render_frame(new*24., update=True)
time_slider.on_trait_change(time_slide, 'value')
```
